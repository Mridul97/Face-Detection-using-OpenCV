# Using an Input Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the face cascade 
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 

# Loading the eye cascade
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
# Loading the image
img = cv2.imread('face.jpg')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# Converting the image to gray
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap = 'gray')
plt.show()
```

```python
# finding faces in the gray image
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

for (x,y,w,h) in faces:
    # Drawing a red rectangle with line width 2 on the image
    cv2.rectangle(img_rgb, (x,y), (x+w, y+h), (255,0,0), 2)

    # Defining the face as the region of image to find the eye in, we will look for eye only on the face
    # Location of the face
    roi_gray = gray[y:y+h, x:x+w]
    # roi in color for later on when we reimpose it
    roi_color = img_rgb[y:y+h, x:x+w]

    # Detecting eyes in the region of interest i.e., the face
    eyes = eye_cascade.detectMultiScale(roi_gray)

    for(ex, ey, ew, eh) in eyes:

        #Drawing rectangle around the eyes on the color image of green color with line width 2
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0,255,0), 2)


plt.imshow(img_rgb)
plt.show()
```

# Using a WebCam

```python
import cv2
import numpy as np

# Loading the face cascade 
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 

# Loading the eye cascade
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

# Loading the webcam
cap = cv2.VideoCapture(0)

while True:
    ret, img = cap.read()
    
    # Converting the image to gray
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # finding faces in the gray image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    for (x,y,w,h) in faces:
        # Drawing a blue rectangle with line width 2 on the image
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)
        
        # Defining the face as the region of image to find the eye in, we will look for eye only on the face
        # Location of the face
        roi_gray = gray[y:y+h, x:x+w]
        # roi in color for later on when we reimpose it
        roi_color = img[y:y+h, x:x+w]
        
        # Detecting eyes in the region of interest i.e., the face
        eyes = eye_cascade.detectMultiScale(roi_gray)
        
        for(ex, ey, ew, eh) in eyes:
            
            #Drawing rectangle around the eyes on the color image of green color with line width 2
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0,255,0), 2)
            
            
    cv2.imshow('img', img)
    k = cv2.waitKey(30) & 0xff
    # if esc key is pressed move out of the loop
    if k == 27:
        break 
        
# Stop using the webcam
cap.release()
cv2.destroyAllWindows()
```
